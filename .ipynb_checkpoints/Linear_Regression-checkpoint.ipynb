```python
# Importing libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Extracting the data and storing it in the dataframe
df = pd.read_csv('ex1data2.txt', names=['SquareFeet', 'Bedrooms', 'HousePrice'])
```

```python
# Plotting the scatterplot between the square feet and the price of the houses
plt.scatter(df.SquareFeet, df.HousePrice)
plt.xlabel("square feet ")
plt.ylabel("House price")
plt.show()
```

```python
# Defining the function to calculate the error for the values m,c
def cost_function(m, c, dataframe):
    total_error = 0

    for i in range(len(dataframe)):
        x = dataframe.iloc[i].SquareFeet
        y = dataframe.iloc[i].HousePrice
        # Calculating the MSE -> Mean Squared Error 
        total_error += (y - (m * x + c)) **2
        
    return total_error / float(len(dataframe))
```

```python
cost_function(0,0,df)
```

```python
# Defining the function to implement the gradient descent
def gradient_descent(m_current, c_current, learning_rate, dataframe):
    n = len(dataframe)
    m_gradient = 0
    c_gradient = 0
     
    # Finding the m, c gradients by the partial derivatives formula
    for i in range(n):
        x = dataframe.iloc[i].SquareFeet
        y = dataframe.iloc[i].HousePrice
        
        m_gradient += ( (2/n) * (y - (m_current * x) - c_current) * (-x) ) 
        c_gradient += ( (2/n) * (y - (m_current * x) - c_current) * (-1) ) 
# Updating the values of m and c
    m_updated = m_current - learning_rate * m_gradient
    c_updated = c_current - learning_rate * c_gradient

    return m_updated, c_updated
```

```python
# Initialising the starting m,c values
m = 0
c = 0
# Learning rate that decides how much we move across the slope
L = 0.00000001
# Number of iterations decided how many times we have to update the value of the m, c
epocs = 500
# Printing the values of m, c and the error
for i in range(epocs):
    m, c = gradient_descent(m, c, L, df)
    print(f"m : {m:.17f} ,c : {c:.17f}, Error : {cost_function(m, c, df):.17f}")
```

```python
# Defining the function that predicts the price of the house based on the input -> square feet 
def linear_regression(m, c, square_feet):
    return m * square_feet + c
```
