```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Importing the data into a dataframe
df = pd.read_csv('ex1data2.txt',names = ['square_feet', 'bedrooms', 'house_price'])
```

```python
# Printing the dataframe
df
```

**Note:**
We can see from the dataframe that the independent features `square_feet` and `bedrooms` are in different scales.If we feed this data into our model, the output will be influenced by the `square_feet` feature because its scale is high as compared to feature `bedrooms`.So, we have to normalise the independent features. 

```python
# Normalising the independent features
independent_features = df[['square_feet', 'bedrooms']]

# Finding the mean and standard deviation for the independent features
mean_value = independent_features.mean()
standard_deviation_value = independent_features.std()

# The features are normalised by subracting the feature value from mean and dividing by standard deviation of the respective feature.
normalised_features = (independent_features - mean_value) / standard_deviation_value

# Adding the normalised features back to the dataframe
df = pd.concat([normalised_features, df['house_price']], axis = 1)
```

```python
# Printing the dataframe after normalisation
df
```

```python
# Defining the cost function
def cost_function(m1, m2, c, dataframe):
    total_error = 0
    n = len(dataframe)
    for i in range(n):
        x1 = dataframe.iloc[i].square_feet
        x2 = dataframe.iloc[i].bedrooms
        y = dataframe.iloc[i].house_price
        # Calculation  of Squared error and returning the MSE -> Mean Squared Error 
        total_error += (y - (m1 * x1) - (m2 * x2) - c) ** 2

    return total_error / n
```

```python
# Defining the function of gradient descent algorithm
def gradient_descent(m1_now, m2_now, c, learning_rate, dataframe):
    m1_gradient = 0
    m2_gradient = 0
    c_gradient = 0
    n = len(dataframe)
    for i in range(n):
        x1 = dataframe.iloc[i].square_feet
        x2 = dataframe.iloc[i].bedrooms
        y = dataframe.iloc[i].house_price
        # Finding the gradients of m1, m2 and c with thier partial derivatives
        m1_gradient += (2 / n) * (( y - (m1_now * x1) - (m2_now * x2) - c ) * -x1)
        m2_gradient += (2 / n) * (( y - (m1_now * x1) - (m2_now * x2) - c ) * -x2)
        c_gradient += (2 / n) * (( y - (m1_now * x1) - (m2_now * x2) - c ) * -1)

    # Updating the values of m1, m2, and c 
    m1_updated = m1_now - (learning_rate * m1_gradient)
    m2_updated = m2_now - (learning_rate * m2_gradient)
    c_updated = c - (learning_rate * c_gradient)

    
    return m1_updated, m2_updated, c_updated
    
    
```

```python
m1 = 0
m2 = 0
c = 0
L = 0.3

epocs = 50

cost = []
iterations = []

for i in range(epocs):
    m1, m2, c = gradient_descent(m1, m2, c, L, df)
    cost.append(cost_function(m1, m2, c, df))
    iterations.append(i)
    print(f"m1 : {m1:.17f}, m2 : {m2:.17f} ,c : {c:.17f}, Error : {cost_function(m1, m2, c, df):.17f}")




print(m1, m2, c)
```

```python
# Plot cost function values over iterations
fig, ax = plt.subplots()  
ax.plot(np.arange(len(iterations)), cost, 'g')  
ax.set_xlabel('Iterations')  
ax.set_ylabel('Cost')  
ax.set_title('Error vs. Training Epoch')  

# Format y-axis ticks to display cost values in regular decimal notation
formatter = plt.FuncFormatter(lambda x, _: '{:,.0f}'.format(x))
ax.yaxis.set_major_formatter(formatter)

plt.show()
```

```python
# Creating the linear regression model to predict the price of the house
def price_prediction_model(square_feet, bedroom_count,m1, m2, c, mean_value, standard_deviation_value):
    # We have to normalise the independent features before prediction
    square_feet = (square_feet - mean_value.iloc[0]) / standard_deviation_value.iloc[0]
    bedroom_count = (bedroom_count - mean_value.iloc[1]) / standard_deviation_value.iloc[1]

    return m1 * square_feet + m2 * bedroom_count + c
```

```python
# Predicted price for 1839 square feet and 2 bedrooms
price_prediction_model(1839,2,m1, m2, c, mean_value, standard_deviation_value)
```

```python
# Predicted price for 2200 square feet and 5 bedrooms
price_prediction_model(2200, 5, m1, m2, c, mean_value, standard_deviation_value)
```

```python
# Predicted price for 1650 square feet and 3 bedrooms
price_prediction_model(1650, 3, m1, m2, c, mean_value, standard_deviation_value)
```
